# Titanic dataset — quick look

Reads `titanic.csv` and makes a simple plot of survival counts by passenger class.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df.head()
```

```python
# Survival count by passenger class
counts = df.groupby(["Pclass", "Survived"]).size().unstack(fill_value=0)
counts.columns = ["Died", "Survived"]

ax = counts.plot(kind="bar", color=["#c0392b", "#27ae60"])
ax.set_xlabel("Passenger class")
ax.set_ylabel("Number of passengers")
ax.set_title("Titanic survival by passenger class")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
